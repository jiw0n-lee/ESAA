### Step 1. Import the necessary libraries

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

### Step 2. Create the DataFrame that should look like the one below.
*참고*

```
raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
            'female': [0, 1, 1, 0, 1],
            'age': [42, 52, 36, 24, 73],
            'preTestScore': [4, 24, 31, 2, 3],
            'postTestScore': [25, 94, 57, 62, 70]}
```

````python
raw_data = {
    'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
    'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
    'age': [42, 52, 36, 24, 73],
    'female': [False, True, True, False, True],
    'preTestScore': [4, 24, 31, 2, 3],
    'postTestScore': [25, 94, 57, 62, 70]
}

df_scores = pd.DataFrame(raw_data)
df_scores.head()
````

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

````python
plt.figure(figsize=(8, 5))
sns.scatterplot(
    x=df_scores['preTestScore'],
    y=df_scores['postTestScore'],
    size=df_scores['age'],
    sizes=(20, 200),
    alpha=0.6
)

plt.title('Pre-Test Score vs Post-Test Score (Bubble Size = Age)')
plt.xlabel('Pre-Test Score')
plt.ylabel('Post-Test Score')
plt.legend(title="Age", loc="upper left", bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
````

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

````python
plt.figure(figsize=(8, 5))
sns.scatterplot(
    x=df_scores['preTestScore'],
    y=df_scores['postTestScore'],
    size=df_scores['postTestScore'] * 4.5,
    hue=df_scores['female'],
    palette={0: 'blue', 1: 'red'},
    sizes=(20, 300),
    alpha=0.6
)

plt.title('Pre-Test Score vs Post-Test Score (Size = 4.5 * PostTestScore, Color = Sex)')
plt.xlabel('Pre-Test Score')
plt.ylabel('Post-Test Score')
plt.legend(title="Sex (0=Male, 1=Female)", loc="upper left", bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
````
